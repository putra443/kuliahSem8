## Pandas Intermediate: Indexing & Slicing<a id="6"></a>


### Memilih kolom
**Ada dua cara untuk memfilter berdasarkan nama kolom:**

**Metode 1:** Cepat dan mudah, tetapi hanya berfungsi jika nama kolom TIDAK memiliki spasi atau karakter khusus.
```python
    df.column_name 
        (returns series)
```

**Metode 2:** Lebih handal dan dapat memfilter beberapa kolom.

```python
    df['column']  
        (returns series)
```

```python 
    df[['column 1', 'column 2']] 
        (returns dataframe)
```
---

````python
df_can.Country  # mengembalikan series
````

# Line Pots (Series/Dataframe) <a id="12"></a>

**Apa Itu Plot Garis dan Mengapa Menggunakannya?**

Bagan garis atau plot garis adalah jenis plot yang menampilkan data sebagai rangkaian titik data yang disebut 'penanda' yang dihubungkan oleh segmen garis lurus. Ini adalah tipe bagan dasar yang umum digunakan di banyak bidang.

Gunakan plot garis jika Anda memiliki kumpulan data berkelanjutan. Diagram ini paling cocok digunakan untuk visualisasi data berbasis tren selama periode waktu tertentu.

**Mari kita mulai dengan studi kasus:**

Pada tahun 2010, Haiti mengalami bencana gempa bumi berkekuatan 7,0. Gempa tersebut menyebabkan kerusakan dan korban jiwa. Sekitar tiga juta orang terkena dampak bencana alam ini. Sebagai bagian dari upaya kemanusiaan Kanada, Pemerintah Kanada meningkatkan upayanya dalam menerima pengungsi dari Haiti. Kita dapat memvisualisasikan upaya ini menggunakan plot `Line`:

**Pertanyaan:** Plot grafik garis imigrasi dari Haiti menggunakan `df.plot()`!

Pertama, kita akan mengekstrak *data series* untuk Haiti.

````python
haiti = df_can.loc['Haiti', years] # mengirimkan kolom years 1980 - 2013 untuk tidak mengikutsertakan kolom 'total'
haiti.head()
````

Selanjutnya, kita akan memplot plot garis dengan menambahkan `.plot ()` ke Dataframe `haiti`.

````python
haiti.plot()
````

*Panda* secara otomatis mengisi sumbu x dengan nilai indeks (tahun), dan sumbu y dengan nilai kolom (populasi). Namun, perhatikan bagaimana tahun tidak ditampilkan karena berjenis *string*. Oleh karena itu, mari kita ubah jenis nilai indeks menjadi *integer* untuk plot.

Selain itu, mari beri label sumbu x dan y menggunakan `plt.title ()`, `plt.ylabel ()`, dan `plt.xlabel ()`:

````python
haiti.index = haiti.index.map(int) # Mari kita ubah nilai indeks Haiti menjadi bilangan bulat untuk plot
haiti.plot(kind='line')

plt.title('Immigration from Haiti')
plt.ylabel('Number of immigrants')
plt.xlabel('Years')

plt.show() # baris ini digunakan untuk menunjukkan pembaruan yang dilakukan pada gambar
````

Kita dapat dengan jelas melihat bagaimana jumlah imigran dari Haiti melonjak dari tahun 2010 ketika Kanada meningkatkan upayanya untuk menerima pengungsi dari Haiti. Mari beri anotasi lonjakan ini di plot dengan menggunakan metode `plt.text ()`.

````python
haiti.plot(kind='line')

plt.title('Immigration from Haiti')
plt.ylabel('Number of Immigrants')
plt.xlabel('Years')

# anotasi gempa 2010
# sintaks: plt.text(x, y, label)
plt.text(2000, 6000, '2010 Earthquake') 

plt.show() 
````

Hanya dengan beberapa baris kode, Anda dapat dengan cepat mengidentifikasi dan memvisualisasikan lonjakan imigrasi!

Catatan singkat tentang nilai x dan y di `plt.text (x, y, label)`:
    
      Karena sumbu x (tahun) adalah tipe 'integer', kita tetapkan x sebagai tahun. Sumbu y (jumlah pendatang) adalah tipe 'integer', jadi kita tinggal menentukan nilai y = 6000.
    
```python
     plt.text (2000, 6000, '2010 Earthquake') # tahun disimpan sebagai tipe int
```
     Jika tahun disimpan sebagai tipe 'string', kita perlu menentukan x sebagai posisi indeks tahun ini. Misal, indeks ke-20 adalah tahun 2000 karena merupakan tahun ke-20 dengan tahun dasar 1980.
`` python
     plt.text (20, 6000, '2010 Earthquake') # tahun disimpan sebagai tipe int
``
     Kita akan membahas metode anotasi lanjutan di modul selanjutnya.

Kita dapat dengan mudah menambahkan lebih banyak negara ke plot garis untuk membuat perbandingan yang bermakna tentang imigrasi dari berbagai negara.

Ingatlah bahwa *Panda* memplot indeks pada sumbu x dan setiap kolom sebagai satu garis pada sumbu y. *Panda* secara otomatis akan membuat grafik kedua negara pada grafik yang sama. Lanjutkan dan plot *dataframe* baru yang sudah di-*transpose*. Pastikan untuk menambahkan judul ke plot dan memberi label pada sumbu.

Plot garis adalah alat visualisasi yang praktis untuk menampilkan beberapa variabel dependen terhadap satu variabel independen. Namun, disarankan tidak lebih dari 5-10 garis pada satu grafik; Jika lebih dari itu, grafik menjadi sulit untuk ditafsirkan.
